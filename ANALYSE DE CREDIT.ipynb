```python
# Detection des outliers
def outliers_detection(df, col):
    IQR = df[col].quantile(0.75) - df[col].quantile(0.25)
    lower = df[col].quantile(0.25) - (1.5*IQR)
    upper = df[col].quantile(0.75) + (1.5*IQR)

    outliers = df.index[(df[col] < lower) | (df[col] > upper)]

    return outliers
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score
```

```python
# On va appliquer 3 algorithms RegressionLogistique, KNN et Arbre de Decision 
models = {
    'LogisticRegression':LogisticRegression(random_state=42),
    'DecisionTreeClassifier':DecisionTreeClassifier(max_depth =1 , random_state=42)
}
#La fonction du taux de bon classement globale(Accuracy)
def accu(y_true,y_pred,retu=False):
    acc =  accuracy_score(y_true,y_pred)
    if retu:
        return acc
    else:
        print(f'Le taux de bon classement du modeles est :{acc}')
#La fonction de la precision
def preci(y_true,y_pred,retu=False):
    pre =  precision_score(y_true,y_pred)
    if retu:
        return pre
    else:
        print(f'La precision du modele est :{pre}')
#La fonction du recall
def recal(y_true,y_pred,retu=False):
    re =  recall_score(y_true,y_pred)
    if retu:
        return re
    else:
        print(f'Le rappel du model est :{re}')
```

```python
#Fonction d'application des modeles
def train_test_eval(models,X_train,y_train,X_test,y_test):
    for name,model in models.items():
        print(name,':')
        model.fit(X_train,y_train)
        accu(y_test,model.predict(X_test))
        preci(y_test,model.predict(X_test))
        recal(y_test,model.predict(X_test))
        print('-'*30)
```

```python
train_test_eval(models,X_train,y_train,X_test,y_test)
```
